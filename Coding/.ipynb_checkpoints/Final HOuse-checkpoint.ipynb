```python
def plotting_3_chart(df, feature):
    import seaborn as sns
    import matplotlib.pyplot as plt
    import matplotlib.gridspec as gridspec
    from scipy import stats
    import matplotlib.style as style
    style.use('fivethirtyeight')
    
    fig = plt.figure(constrained_layout =True, figsize=(12,8)) #create chart
    grid = gridspec.GridSpec(ncols=3, nrows=3, figure=fig) # 3cols and 3rows
    
    ax1 = fig.add_subplot(grid[0, :2])
    ax1.set_title('Histogram') # title
    sns.distplot(df.loc[:,feature], norm_hist=True, ax =ax1)
    
    ax2 = fig.add_subplot(grid[2, :2])
    ax2.set_title('2nd plot')
    stats.probplot(df.loc[:,feature], plot = ax2)
    
    
    ax3 = fig.add_subplot(grid[:,2])
    ax3.set_title('Box Plot')
    sns.boxplot(df.loc[:,feature], orient='v', ax = ax3);
```

```python
# get normal distribution 
df["SalePrice"] = np.log1p(df["SalePrice"])
plotting_3_chart(df, 'SalePrice')
```

```python
X = final.drop('SalePrice', axis=1) #from train
y = final['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", X_test.shape)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# menually created model checking sysytem !!
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+'-->'+str(acc)) #acc = accuracy
```

```python
#LinearRegressio - 1st accuracy check

from sklearn.linear_model import LinearRegression # (lr = LinearRegression)
lr = LinearRegression()
model_acc(lr)
```

```python
# 3rd accuracy check with DecisionTreeRegressor

from sklearn.tree import DecisionTreeRegressor 
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
# 4th accuracy check with RandomForestRegressor 

from sklearn.ensemble import RandomForestRegressor  
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
             'criterion':['squared_error','absolute_error','poisson']} 
grid_obj = GridSearchCV(estimator=rf, param_grid=parameters) # rf = RandomForestRegressor

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_
best_model

# absolute_error is the best model
```

```python
best_model.score(X_test , y_test)
```
